```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/deep learning/train.csv')
```

```python
df.head()
```

```python
df=df.dropna()
```

```python
x=df.drop('label',axis=1)
```

```python
y=df['label']
```

```python
x.shape
y.shape
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size=5000
```

```python
messages=x.copy()
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0,len(messages)):
  print(i)
  review=re.sub('[^a-zA-Z]',' ',messages['title'][i])
  review=review.lower()
  review=review.split()

  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
# creating model
embedding_vectors_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vectors_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
x_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
x_final.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.3,random_state=42)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

```python
# The threshold in binary classification is a crucial parameter that determines
# how predicted probabilities are translated into class labels
# . It defines the boundary between considering a sample as belonging to the positive
# class or the negative class. When the predicted probability for a sample exceeds
# the threshold, it's classified as the positive class; otherwise, it's classified
# as the negative class.

import numpy as np
from sklearn.metrics import confusion_matrix

# Assuming 'model' is your trained Sequential model
y_pred_prob = model.predict(x_test)

# Threshold the predicted probabilities to get binary class predictions
threshold = 0.5  # You can adjust this threshold as needed
y_pred_binary = (y_pred_prob >= threshold).astype(int)

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred_binary)

print(cm)

```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_binary)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_binary))
```
